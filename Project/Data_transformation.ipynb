```python
import pandas as pd
import numpy as np
from termcolor import colored
from sklearn.preprocessing import LabelEncoder
```

```python
train_values = pd.read_csv('./Datasets/train_values.csv')
train_labels = pd.read_csv('./Datasets/train_labels.csv')

print("\n Training Data \n ")

train_values.head()
```

```python
test_values = pd.read_csv("./Datasets/test_values.csv")
print(colored("\n Testing Data \n ",'blue'))
test_values.head()
```

```python
train_rows,train_cols = train_values.shape
test_rows,test_cols = test_values.shape
print(f"Training\n  Rows : {train_rows} | Columns : {train_cols}")
print(f"Testing \n  Rows : {test_rows}  | Columns : {test_cols}")
```

```python
print("Training Labels")
train_labels.head()
```

### Combining Dataset
**Combining the features and labels for shuffling**<br>
**Remove building_id coulumn to misguide the model**

```python
train_values['damage_grade'] = train_labels['damage_grade']

combined_df = train_values.drop(columns=['building_id'])
combined_df.head()
```

### Convert Categorical features into Numerical features

```python
label_encoder = LabelEncoder()

categorical = [f for f in combined_df.columns if combined_df.dtypes[f] == 'object']

for col in categorical:
        combined_df[col] = label_encoder.fit_transform(combined_df[col])

encoded_df = combined_df

categorical = [f for f in encoded_df.columns if encoded_df.dtypes[f] == 'object']

if len(categorical)!=0:
     print("Categorical Values found before training!!")
else:
    print(colored("No Categorical Features Found -ready-for training",'blue'))
```

```python
from sklearn.preprocessing import StandardScaler
 
scaler = StandardScaler()
scaled_data = scaler.fit_transform(encoded_df.drop(columns=['damage_grade']))
damage_grade = encoded_df['damage_grade']
scaled_df = pd.DataFrame(scaled_data,
                         columns=encoded_df.columns[:-1])
scaled_df['damage_grade'] = damage_grade
display(scaled_df.head())
```

#### Saving the preprocessed Datasets

```python
encode_df.to_csv("./Datasets/encoded_df.csv")
scaled_df.to_csv("./Datasets/Scaled_df.csv",index=False)
```
